# 1.label Encoding

```python
data = ['None','low','high','very high','medium']
data = pd.DataFrame(data,columns=['types'])
```

```python
data
```

```python
#create object of labelencoder
labelencoder = LabelEncoder()
label_data = labelencoder.fit_transform(data['types'])
label_data
```

```python
#create column of label_data
data['label'] = label_data
data
```

```python
#convert label column into categorical variable
category = labelencoder.inverse_transform(data['label'])
category
```

```python
#make category as the column
data['conver_type'] = category
data
```

# OneHotEncoder


In this strategy, each category value is converted into a new column and assigned a 1 or 0 (notation for true/false) value to the column

In this technique, the categorical parameters will prepare separate columns for both Male and Female labels. So, wherever there is Male, the value will be 1 in Male column and 0 in Female column, and vice-versa.
