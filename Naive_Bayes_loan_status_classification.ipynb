Fill all null value with his mean value using for loop

```python
ls = ['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6','PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']
```

```python
min_threshold,max_threshold = data.AGE.quantile([0.001,0.999])
min_threshold,max_threshold
```

```python
data[data.AGE < min_threshold]
```

```python
data = data[data.AGE<max_threshold]
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
model = LabelEncoder()
data['Default Status']=model.fit_transform(data['Default Status'])
```

```python
data['Default Status']
```

```python
data = data.drop(['ID'],axis=1)
```

```python
data
```

```python
#dependent and independent variable
x=data.drop(['Default Status'],axis=1)
x
```

```python
y=data['Default Status']
```

```python
#split the  data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
