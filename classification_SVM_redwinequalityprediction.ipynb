```python
#create target and feature variable
x=data.drop("Quality_Category",axis="columns")
```

```python
y = data['Quality_Category']
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
#split data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 10)
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
#checing overfitting and under fitting
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
#check accuracy
from sklearn.metrics import accuracy_score,precision_score,recall_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
model_new = SVC(C=100)
model_new.fit(x_train,y_train)
```

```python
y_predict_new= model_new.predict(x_test)
```

```python
accuracy_score(y_test,y_predict_new)
```
