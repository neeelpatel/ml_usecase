# feature scaling

# 1.Normalization(Min_Max Normalization)

```python
data.describe()
```

```python
scaling = MinMaxScaler()
```

```python
data[['Pregnancies_','Glucose_']] = scaling.fit_transform(data[['Pregnancies','Glucose']])
```

```python
data
```

```python
#ploting
plt.figure(figsize = (10,5))
plt.hist(data.Pregnancies)
```

```python
plt.figure(figsize = (10,5))
plt.hist(data.Pregnancies_)
```

```python
plt.hist(data.Glucose)
```

```python
plt.hist(data.Glucose_)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize =(20, 5)) 

sns.kdeplot(data['Pregnancies'], ax = ax1, color ='r') 
sns.kdeplot(data['Glucose'], ax = ax1, color ='b') 
ax1.set_title('Before Scaling') 
  
sns.kdeplot(data['Pregnancies_'], ax = ax2, color ='red') 
sns.kdeplot(data['Glucose_'], ax = ax2, color ='blue') 
ax2.set_title('After Min-Max Scaling') 
plt.show() 
```

# standardization

Standardization is a scaling technique where the values are centered around the mean with a unit standard deviation. This means that the mean of the attribute becomes zero and the resultant distribution has a unit standard deviation.


```python
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scale = StandardScaler()
```

```python
data[['BloodPressure_','SkinThickness_']] =  std_scale.fit_transform(data[['BloodPressure','SkinThickness']])
```

```python
data[['BloodPressure_','SkinThickness_']]
```

```python
data['BloodPressure'].mean()
```

#data

```python
print(data['BloodPressure_'].min(),data['BloodPressure_'].max())
```

```python
print(data['SkinThickness_'].min(),data['SkinThickness_'].max())
```

# RobustScaler

 If there are outliers, use RobustScaler(). Alternatively you could remove the outliers and use either of the above 2 scalers (choice depends on whether data is normally distributed)

```python
from sklearn.preprocessing import RobustScaler
```

```python
scale_Robust = RobustScaler()
```

```python
data[['Insulin_','BMI_']] = scale_Robust.fit_transform(data[['Insulin','BMI']])
```

```python
data
```

# Scaling through Visualization

# With and Without Min Max Scaler

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,6))
ax1.set_title('Before Scaling') 
sns.kdeplot(data['Pregnancies'],ax=ax1,color='b')
sns.kdeplot(data['Glucose'],ax=ax1,color='r')
ax2.set_title('after Scaling') 

sns.kdeplot(data['Pregnancies_'],ax=ax2,color='b')
sns.kdeplot(data['Glucose_'],ax=ax2,color='r')
plt.show()
```

# With and Without  standardization


```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,6))
ax1.set_title('before scaling')
sns.kdeplot(data['BloodPressure'],ax=ax1,color='b')
sns.kdeplot(data['SkinThickness'],ax=ax1,color='r')

sns.kdeplot(data['BloodPressure_'],ax=ax2,color='b')
sns.kdeplot(data['SkinThickness_'],ax=ax2,color='r')
ax2.set_title('after scaling')
```

# with and without robust scaling

```python
fig,(ax1,ax2)= plt.subplots(ncols=2,figsize=(12,6))
ax1.set_title('before scaling')
sns.kdeplot(data['Insulin'],ax=ax1,color='b')
sns.kdeplot(data['BMI'],ax=ax1,color='r')
sns.kdeplot(data['Insulin_'],ax=ax2,color='b')
sns.kdeplot(data['BMI_'],ax=ax2,color='r')
ax2.set_title('after scaling')
```
